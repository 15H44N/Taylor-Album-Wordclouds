```python
import pandas as pd
```

```python
alllyricsdf=pd.read_csv("./taylor_swift_lyrics.csv",encoding='windows-1252')
alllyricsdf.drop(columns=["track_n","line","artist"],inplace=True)
alllyricsdf.head()
```

```python
import os
album_const ="Lover"
year_const = 2020

directory = r'C:/Users/ishaa/Files/Notebook Projects/Miss Americana/Lyrics/LoverLyrics'

song_lyric_dict={}

for filename in os.listdir(directory):
    if(filename.endswith(".txt")):
        songname=filename.split(".txt")[0]
        #print(songname)
        #song_lyric_dict[songname]
    
        with open(directory+"/" + filename) as f:
            file_lines=f.readlines()
    
        song_lyric_dict[songname] = file_lines
```

```python
data_dict={}
for i in song_lyric_dict.items():
    song_name=i[0]
    val=i[1]
    for j in val:
        data_dict[j.split("\n")[0]] = {"track_title":song_name , "year":year_const , "album":album_const}
```

```python
final_data_list=[]
for i in data_dict.items():
    lyric_item=i[0]
    infodict =i[1]
    infodict["lyric"] = lyric_item
    final_data_list.append(infodict)    
```

```python
lover_lyric_df=pd.DataFrame(final_data_list)
lover_lyric_df = lover_lyric_df[['album', 'track_title', 'lyric', 'year']]
lover_lyric_df
```

```python
final_df = alllyricsdf.append(lover_lyric_df)
final_df.reset_index(inplace=True)
final_df.drop(columns="index",inplace=True)
final_df
```

```python
final_df.to_csv("./FinalTaylorSwiftLyrics.csv",index=False)
```
